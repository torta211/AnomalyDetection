```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
NAB_keys = []
for x in glob.glob("NAB/results/NumentaTM/*/*.csv"):
    f = x.split("\\")[-1].split("numentaTM_")[-1]
    fo = x.split("\\")[1]
    NAB_keys.append(f"{fo}/{f}")
```

```python
NAB_keys
```

```python
dsets = {}
for k in NAB_keys:
    dsets[k] = pd.read_csv(f"data_results/slidingEnsemble/{k[:-4]}_dataset45.csv").drop(["Unnamed: 0"], axis="columns")
```

```python
dsets[NAB_keys[0]]
```

```python
plt.figure(figsize=(17, 7))
plt.plot(dsets[NAB_keys[0]]["val"])
plt.twinx().plot(dsets[NAB_keys[0]]["min_values"], "orange")
```
